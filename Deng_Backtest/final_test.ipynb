demo使用教程：api函数为run_backtest(strategy, start_date=None, end_date=None, stock_pool='000852.XSHG', data_fields=None)\
1.参数strategy可以使用已经封装好的n日反转策略类ReversalStrategy和双均线策略MAStrategy。当然，如果你想改一下策略的参数（例如n日的n），请于test_api函数中找到两个类的定义，然后修改参数即可。\
2.关于1的补充,事实上，该回测框架支持几乎所有策略的回测，故你大可以自己写一个Mystrategy类进行回测，随意设置你所需要的参数，下面将在demo3进行演示。\
3.参数start_date和参数end_date是策略回测的起始时间，你需要传入一个形如datetime(2020,1,2)的时间（请注意，不要传入老师所给数据时间范围以外的数据，即2020-01-02之前和2023-12-29之后）\
4.参数stock_pool为股票池，现暂时设定为指数（上证综指，沪深300，中证500，中证800，中证1000），分别对应['000001.XSHG','000300.XSHG','000905.XSHG','000906.XSHG','000852.XSHG']，股票池意味着策略所使用资产池以及最终对比的benchmark。\
5.参数data_fields表示选择哪些数据用于回测，为了简便和数据的轻量化，我们只开放了OHLCV数据，你可以自由选择你的策略要用到这五个数据中的哪些。（请注意，该参数接受一个列表，列表中的元素的顺序不做要求）

！！！下面是自定义策略的回测，也是回测框架最核心部分，此处我们回测WorldQuant016因子(窗口区间取5)，其定义为：
$Alpha016=-rank(cov(rank(high),rank(volume)),5)$

```python
#demo3
##策略类定义
class MyStrategy(Strategy):
    """高价量背离策略"""
    def __init__(self, broker, data, n_days=5):
        super().__init__(broker, data)
        self.n_days = n_days  # N日周期
        self.factor_values = {}  # 每只股票的因子值
        self.buy_signals = []  # 存储产生买入信号的股票信息
        super().__setattr__('broker', broker)
        
    def next(self):
        """策略逻辑"""
        self.buy_signals = []  # 清空买入信号列表
        
        # 清空昨日持仓
        for stock_code in list(self.broker.positions.keys()):
            size = self.broker.positions[stock_code]
            if size > 0:
                close_price = self.data[stock_code].lines['close'][-1]
                self.sell(size, close_price, stock_code)
        
        # 计算每只股票的因子值
        factor_dict = {}
        for stock_code, data in self.data.items():
            # 获取最高价和成交量序列
            high_prices = list(data.lines['high'])
            volumes = list(data.lines['volume'])
            
            # 确保有足够的数据计算因子
            if len(high_prices) >= self.n_days and len(volumes) >= self.n_days:
                # 获取最近N日数据
                recent_highs = high_prices[-self.n_days:]
                recent_vols = volumes[-self.n_days:]
                
                # 计算rank序列
                high_ranks = pd.Series(recent_highs).rank()
                vol_ranks = pd.Series(recent_vols).rank()
                
                # 计算rank的协方差并取负
                factor_value = -high_ranks.cov(vol_ranks)
                factor_dict[stock_code] = factor_value
        
        # 选出因子值最大的前20%的股票
        if factor_dict:
            sorted_stocks = sorted(factor_dict.items(), key=lambda x: x[1], reverse=True)
            num_stocks = max(1, int(len(sorted_stocks) * 0.2))
            selected_stocks = sorted_stocks[:num_stocks]
            
            # 生成买入信号
            for stock_code, _ in selected_stocks:
                self.buy_signals.append({
                    'stock_code': stock_code,
                    'price': self.data[stock_code].lines['close'][-1]
                })
        
        # 处理买入信号
        if self.buy_signals:
            available_cash = self.broker.cash * 0.8  # 可用资金为总资金的80%
            per_stock_cash = available_cash / len(self.buy_signals)  # 每只股票分配的资金
            
            # 执行买入订单
            for signal in self.buy_signals:
                size = int(per_stock_cash / signal['price'])  # 计算可买入的股票数量
                if size > 0:
                    self.buy(size, signal['price'], signal['stock_code'])
```

```python
#demo3续
##该demo演示了自定义策略（单因子策略）的回测
back = run_backtest(MyStrategy, datetime(2021,1,2), datetime(2021,12,30), '000300.XSHG', ['open', 'high', 'low', 'close', 'volume'])
```
